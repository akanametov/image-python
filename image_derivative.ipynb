```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
from scipy.ndimage import filters
```

Derivative with **Sobel** filter

```python
img = Image.open("images/empire.jpg").convert('L') # Convert to grayscale

I = np.array(img)/255

Ix = filters.sobel(I, axis=1)
Iy = filters.sobel(I, axis=0)

M = np.sqrt(Ix**2 + Iy**2).astype(I.dtype) # Magnitude
```

Visualizing image and its derivatives

```python
fig, (ax1, ax2, ax3, ax4) = plt.subplots(1, 4, figsize=(16, 5))
fig.suptitle('Result of "sobel" filter')

ax1.imshow(I, cmap='gray')
ax1.set_title('Original image')

ax2.imshow(Ix, cmap='gray')
ax2.set_title('Derivative in x-dimension (width)')

ax3.imshow(Iy, cmap='gray')
ax3.set_title('Derivative in y-dimension (height)')

ax4.imshow(M, cmap='gray')
ax4.set_title('Magnitude of derivatives')
plt.show()
```

Derivative with **Gaussian** filter

```python
sigma = 5

Ix = filters.gaussian_filter(I, sigma=(sigma, sigma), order=(0,1))
Iy = filters.gaussian_filter(I, sigma=(sigma, sigma), order=(1,0))

M = np.sqrt(Ix**2 + Iy**2).astype(I.dtype) # Magnitude
```

```python
fig, (ax1, ax2, ax3, ax4) = plt.subplots(1, 4, figsize=(16, 5))
fig.suptitle('Result of "gaussian_filter"')

ax1.imshow(I, cmap='gray')
ax1.set_title('Original image')

ax2.imshow(Ix, cmap='gray')
ax2.set_title('Derivative in x-dimension (width)')

ax3.imshow(Iy, cmap='gray')
ax3.set_title('Derivative in y-dimension (height)')

ax4.imshow(M, cmap='gray')
ax4.set_title('Magnitude of derivatives')
plt.show()
```
