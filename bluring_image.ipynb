```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
from scipy.ndimage import filters
```

Blur **grayscale** image

```python
img = Image.open("images/empire.jpg").convert('L') # Convert to grayscale
img = np.array(img)
new_img = filters.gaussian_filter(img, sigma=5)
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2)
fig.suptitle('Result of "gaussian_filter"')

ax1.imshow(img, cmap='gray')
ax1.set_title('Original image')

ax2.imshow(new_img, cmap='gray')
ax2.set_title('Blurred image')
plt.show()
```

Blur **RGB** image

```python
def blur_rgb(img, sigma=5):
    assert img.ndim == 3, 'image should be represented as 3-dimensional tensor'
    C = img.shape[-1] # number of image channels (R, G, B)
    new_img = np.zeros(img.shape)
    for c in range(C): # for each channel apply gaussian_filter(...)
        new_img[..., c] = filters.gaussian_filter(img[..., c], sigma=sigma)
    return new_img.astype(img.dtype) # convert to input dtype
```

```python
img = Image.open("images/empire.jpg").convert('RGB') # Convert to RGB
img = np.array(img)
new_img = blur_rgb(img, sigma=5)
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2)
fig.suptitle('Result of "blur_rgb"')

ax1.imshow(img)
ax1.set_title('Original image')

ax2.imshow(new_img)
ax2.set_title('Blurred image')
plt.show()
```
